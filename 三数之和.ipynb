```python
def threeSum(nums):
    if len(nums) < 3:
        return []
    elif len(set(nums)) == 1 and nums[0] == 0:
        return [nums[:3]]
    List = []
    positive = [i for i in nums if i > 0]
    negtive = [abs(i) for i in nums if i < 0]
    zero = [i for i in nums if i == 0]
    a = -1
    for i in positive:
        a += 1
        for j in positive[a+1:]:
            if i+j in negtive:
                if sorted([i, j, -(i+j)]) not in List:
                    List.append(sorted([i, j, -(i+j)]))
    b = -1
    for i in negtive:
        b += 1
        for j in negtive[b+1:]:
            if i+j in positive:
                if sorted([-i, -j, i+j]) not in List:
                    List.append(sorted([-i, -j, i+j]))
    if len(zero) > 2:
        List.append([0, 0, 0])
    if len(zero) > 0:
        both = set(positive).intersection(set(negtive))
        for i in both:
            List.append([-i, 0, i])
    return List
```

排序+双指针：


```python
def threeSum(nums):
    n=len(nums)
    if(not nums or n<3):
        return []
    nums.sort()
    res=[]
    for i in range(n):
        if(nums[i]>0):
            return res
        if(i>0 and nums[i]==nums[i-1]):
            continue
        L=i+1
        R=n-1
        while(L<R):
            if(nums[i]+nums[L]+nums[R]==0):
                res.append([nums[i],nums[L],nums[R]])
                while(L<R and nums[L]==nums[L+1]):
                    L=L+1
                while(L<R and nums[R]==nums[R-1]):
                    R=R-1
                L=L+1
                R=R-1
            elif(nums[i]+nums[L]+nums[R]>0):
                R=R-1
            else:
                L=L+1
    return res
```

```python
nums = [-4,-2,1,-5,-4,-4,4,-2,0,4,0,-2,3,1,-5,0]
threeSum(nums)
```
