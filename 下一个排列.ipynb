* ******** (i,j) A[i] < A[j] [j,end) 
*  [j,end) **** A[i] < A[k]  kA[i]A[k] 
*  A[i]  A[k] 
*  [j,end)  [j,end)
*  1  [begin,end)  4

*** (i,j) i=4j=5 57***

*** [j,end)  A[i]  A[k] A[i]  5 A[k]  6***

*** A[i]  A[k]  56***

*** [j,end)  [j,end) [7,5,4]***

******


```python
class Solution:
    def nextPermutation(self, nums) -> None:
        n = len(nums)
        if n<2: return nums
        i = n-1
        while i>0 and nums[i-1]>=nums[i]:#    
            i -= 1
        if i==0 : # i==0 and nums[i]==max(nums)
            return nums[::-1]
        else:
            j = n-1
            while j>i-1 and nums[j]<=nums[i-1]:
                j -= 1
            nums[i-1], nums[j] = nums[j], nums[i-1]
            nums[i:] = sorted(nums[i:])
        return nums
```

```python
solve = Solution()
list = [1, 5, 2, 8, 6]
res = solve.nextPermutation(list)
print(res)
```
