```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def generateList(l: list) -> ListNode:
    prenode = ListNode(0)
    lastnode = prenode
    for val in l:
        lastnode.next = ListNode(val)
        lastnode = lastnode.next
    return prenode.next

def printList(l: ListNode):
    while l:
        print("%d" %(l.val), end = '')
        l = l.next
    print('')
```

## 优先级队列
建立优先队列，不需要全部元素一次性入队；只需要让链表头元素入队即可，弹出该元素后，该链表往后移。
### 复杂度分析
* 空间复杂度 $\mathcal{O}(k)$
* 时间复杂度 $\mathcal{O}(nlogk)$，n是所有节点个数，k是链表数

```python
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        import heapq
        dummy = ListNode(0)
        p = dummy
        head = []
        for i in range(len(lists)):
            if lists[i] :
                heapq.heappush(head, (lists[i].val, i))
                lists[i] = lists[i].next
        while head:
            val, idx = heapq.heappop(head)
            p.next = ListNode(val)
            p = p.next
            if lists[idx]:
                heapq.heappush(head, (lists[idx].val, idx))
                lists[idx] = lists[idx].next
        return dummy.next
```

## 分治合并合并
* k个链表两两配对，进行第一轮合并，结束后k个链表被合并成k/2个链表
* k/2个链表依然两两配对，进行第二轮合并，结束后k/2个链表被合并成k/4个链表
* 重复上述过程，进行$\mathcal{log_2k}$次合并，完成总体合并工作


### 复杂度分析
* 时间复杂度 $\mathcal{O}(n*log_2k)$ 。K个有序链表，每次分割一半，总共需要分割$\mathcal{log_2k}$次，而每一轮merging的时间复杂度为$\mathcal{O}(n)$，注意这里的$\mathcal{n}$是所有链表的节点总数，因为每一次merging所有的节点都需要比较一次，可参考上述分治合并过程的图片。则时间复杂度求和为：$\mathcal{log_2k}$ * $\mathcal{O}(n)$ = $\mathcal{O}(n*log_2k)$
* 空间复杂度 $\mathcal{O}(1)$

```python
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        amount = len(lists)
        interval = 1
        while interval < amount:
            for i in range(0, amount - interval, interval * 2):
                lists[i] = self.merge2Lists(lists[i], lists[i + interval])
            interval *= 2
        return lists[0] if amount > 0 else lists

    def merge2Lists(self, l1, l2):
        head = point = ListNode(0)
        while l1 and l2:
            if l1.val <= l2.val:
                point.next = l1
                l1 = l1.next
            else:
                point.next = l2
                l2 = l1
                l1 = point.next.next
            point = point.next
        if not l1:
            point.next=l2
        else:
            point.next=l1
        return head.next
```

## 递归
先将所有要合并的链表转换为列表进行排序，再把此列表转换为链表即可

```python
class Solution:
    def cover(self, a:ListNode):
        b = []
        while a:
            b.append(a.val)
            a = a.next
        return b
    def mergeKLists(self, lists) -> ListNode:
        S = []
        for i in range(len(lists)):
            b = self.cover(lists[i])
            S.append(b) 
        S1 = []
        for i in range(len(S)):
            for j in S[i]:
                S1.append(j)
        if len(S1) == 0:
            return []
        S1.sort()
        first = S1.pop(0)
        result = ListNode(first)
        node = result
        for i in S1:
            node.next = ListNode(i)
            node = node.next
        return result
```
