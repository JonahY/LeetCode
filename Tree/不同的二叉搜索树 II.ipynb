```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def generateTrees(self, n: int):
        if(n==0):
            return []
        def build_Trees(left,right):
            all_trees=[]
            if(left>right):
                return [None]
            for i in range(left,right+1):
                left_trees=build_Trees(left,i-1)
#                 print('l', left_trees[0], left, right)
                right_trees=build_Trees(i+1,right)
#                 print('r', right_trees[0], left, right)
                for l in left_trees:
                    print('l', l, i)
                    for r in right_trees:
                        print('r', r, i)
                        cur_tree=TreeNode(i)
                        cur_tree.left=l
                        cur_tree.right=r
                        draw(cur_tree)
                        all_trees.append(cur_tree)
            return all_trees
        res=build_Trees(1,n)
        return res
    
    def create_graph(self, G, node, pos={}, x=0, y=0, layer=1):
        pos[node.val] = (x, y)
        if node.left:
            G.add_edge(node.val, node.left.val)
            l_x, l_y = x - 1 / 2 ** layer, y - 1
            l_layer = layer + 1
            create_graph(G, node.left, x=l_x, y=l_y, pos=pos, layer=l_layer)
        if node.right:
            G.add_edge(node.val, node.right.val)
            r_x, r_y = x + 1 / 2 ** layer, y - 1
            r_layer = layer + 1
            create_graph(G, node.right, x=r_x, y=r_y, pos=pos, layer=r_layer)
        return (G, pos)

    def draw(self, node):   # 以某个节点为根画图
        graph = nx.DiGraph()
        graph, pos = create_graph(graph, node)
        fig, ax = plt.subplots(figsize=(5, 7))  # 比例可以根据树的深度适当调节
        nx.draw_networkx(graph, pos, ax=ax, node_size=30)
        plt.show()
```
