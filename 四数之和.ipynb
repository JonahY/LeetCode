```python
import numpy as np
```

```python
def fourSum(nums, target):
    n=len(nums)
    if(not nums or n<4):
        return []
    nums.sort()
    res=[]
    for j in range(n-3):
        if(nums[j]+nums[j+1]+nums[j+2]+nums[j+3]>target):
            break
        if(nums[j]+nums[-1]+nums[-2]+nums[-3]<target):
            continue
        if(j>0 and nums[j]==nums[j-1]):
            continue
        for i in range(j+1, n-2):
            if(nums[j]+nums[i]+nums[i+1]+nums[i+2]>target):
                break
            if(nums[j]+nums[i]+nums[-1]+nums[-2]<target):
                continue
            if(i>j+1 and nums[i]==nums[i-1]):
                continue
            L=i+1
            R=n-1
            while(L<R):
                if(nums[j]+nums[i]+nums[L]+nums[R]==target):
                    res.append([nums[j],nums[i],nums[L],nums[R]])
                    while (L<R and nums[L]==nums[L+1]):
                        L=L+1
                    while (L<R and nums[R]==nums[R-1]):
                        R=R-1
                    L=L+1
                    R=R-1
                elif nums[j]+nums[i]+nums[L]+nums[R]>target:
                    R=R-1
                else:
                    L=L+1
    return res
```

```python
nums = [-1,0,1,2,-1,-4]
target = -1
fourSum(nums, target)
```
